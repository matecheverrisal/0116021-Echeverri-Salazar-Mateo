```python
#20
import numpy as np

def dimensiones():
    while True:
        try:
            m = int(input("Ingrese el número de filas (m): "))
            n = int(input("Ingrese el número de columnas (n): "))
            assert m >= 0 and n >= 0
            break
        except (ValueError, TypeError):
            print("Ha ingresado un valor no válido. Intente de nuevo.")
            continue

    return m, n

def crea_matriz(m, n):
    matriz = [[0 for c in range(n)] for f in range(m)]
    for fila in range(len(matriz)):
        for elemento in range(len(matriz[fila])):
            while True:
                try:
                    inp = input(f"Ingrese el valor de la posición "
                                f"{chr(65 + elemento)}{fila + 1}: ")
                    matriz[fila][elemento] = int(inp)
                    break
                except (ValueError, TypeError):
                    print("Ha ingresado un valor no válido."
                              "Intente de nuevo.")
                    continue

    return matriz

def mi_rot90(matriz):
    for i , fila in enumerate(matriz):
        matriz[i] = fila[::-1]

    rot90 = matriz.T

    return rot90
m, n = dimensiones()
matriz = np.array(crea_matriz(m, n))

print("Matriz original: ")
print(matriz)
print("Matriz rotada 90°:")
print(mi_rot90(matriz))
```
