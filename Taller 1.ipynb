```python
#9
x1 =abs(int(input("ingrese la coordenada x del circulo 1: ")))
y1 =abs(int(input("ingrese la coordenada y del circulo 1: ")))
r1 =abs(int(input("ingrese el radio del circulo 1: ")))
x2 =abs(int(input("ingrese la coordenada x del circulo 2: ")))
y2 =abs(int(input("ingrese la coordenada y del circulo 2: ")))
r2 =abs(int(input("ingrese el radio del circulo 2: ")))
x =abs(int(input("ingrese la coordenada x del punto: ")))
y =abs(int(input("ingrese la coordenada y del punto: ")))
dist_r1_punto =((x - x1)**2+(y-y1)**2)**(0.5)
dist_r2_punto =((x - x2)**2+(y-y2)**2)**(0.5)
if (dist_r1_punto<=r1)and(dist_r2_punto<= r2):
    print("El punto evaluado se encuentra en los circulos")
elif (dist_r1_punto>r1)and(dist_r2_punto> r2):
    print("El punto evaluado se encuentra fuera de los circulos")
else:
    if (dist_r1_punto<=r1):
        print("El punto evaluado se encuentra en el circulo 1")
    if (dist_r2_punto<=r2):
        print("El punto evaluado se encuentra en el circulo 2")
```

```python
#10
l1 = list(input("Digite el texto: "))
largo= len(l1)
largo1=largo
l2 =[]
while largo>0:
    elemento=l1[largo1-largo]
    posicion=ord(elemento)
    posicion_nueva=posicion-32
    nuevo_elemento=chr(posicion_nueva)
    l2.append(nuevo_elemento)
    largo-=1
k= "".join(l2)
print("el texto digitado en mayuscula es: ",k)
```
